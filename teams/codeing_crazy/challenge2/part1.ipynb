Setting the signal to low when it is already low causes a zero-width spike, and these spikes are aligned across the two strings.

The top signal is basically a clock, so we can ignore it.

The bottom signal seems to have useful data.

We will work with these as binary strings to more easily find patterns. This will let us convert them to characters in a reasonably easy way.

Once we remove the sections with the spikes in the clock (the clock stutters there as well), we can see that string2 is sending data at a rate of 1 bit of information per 4 clock cycles.

Interpreting these bits as ASCII gives us a website.

```python
import binascii

string2_bin = ''
bit = 0

for i in range(len(string2)):
    if string2[i] == "h":
        bit = 1
    if string2[i] == "l":
        bit = 0
    string2_bin += str(bit)
    
#print(string2_bin)

string2_bin_fix = ''#string2_bin[0:4]
ind = 4
while ind <= len(string2_bin):
    string2_bin_fix+=string2_bin[ind:ind+32]
    ind+=46
    
#print(len(string2_bin_fix))
string2_bin_fix = '0b'+string2_bin_fix
n = int(string2_bin_fix,2)
#print(binascii.unhexlify('%x' % n))

#print(string2_bin_fix)

ind = 2
string3 = ''
while ind < len(string2_bin_fix):
    if int(string2_bin_fix[ind:ind+4]) != 0:
        string3 += '1'
    else:
        string3 += '0'
    ind+=4
    
#print(string3)
m = int(string3[2:],2)
string4 = str(binascii.unhexlify('%x' % m))
print(string4[3:len(string4)-1])
```
